```python
"""
Carrega base de dados
"""

raw_df = spark.read.format('delta').table('workspace.pratica.churn_modelling')
```

```python
"""
Renomeia colunas do dataframe
"""

mapeamento_rename = {
    "RowNumber":        "id_registro",
    "CustomerId":       "id_cliente",
    "Surname":          "sobrenome",
    "CreditScore":      "score_credito",
    "Geography":        "pais",
    "Gender":           "genero",
    "Age":              "idade",
    "Tenure":           "anos_relacionamento",
    "Balance":          "saldo",
    "NumOfProducts":    "qtd_produtos",
    "HasCrCard":        "possui_cartao_credito",
    "IsActiveMember":   "membro_ativo",
    "EstimatedSalary":  "salario_estimado",
    "Exited":           "flag_saiu",
}

renamed_df = raw_df

for old_name, new_name in mapeamento_rename.items():
    renamed_df = renamed_df.withColumnRenamed(old_name, new_name)

renamed_df.printSchema()
```
