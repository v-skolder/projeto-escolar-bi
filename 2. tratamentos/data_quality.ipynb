```python
"""
Importa dependências
"""

import pyspark.sql.functions as F
```

```python
"""
Define funções
"""

def get_null_analysis(df, target_column_name):
    total_count = df.count()
    null_count = df.filter(f'{target_column_name} is null').count()
    taxa_null = round(null_count / total_count * 100, 2)

    return {
        'null_count': null_count,
        'taxa_null': taxa_null
    }

def get_negative_analysis(df, target_column_name):
    total_count = df.count()
    negative_count = df.filter(f'{target_column_name} < 0').count()
    taxa_negative = round(negative_count / total_count * 100, 2)

    return {
        'negative_count': negative_count,
        'taxa_negative': taxa_negative
    }

def get_empty_string_analysis(df, target_column_name):
    total_count = df.count()
    empty_string_count = df.filter(f"{target_column_name} = ''").count()
    taxa_empty_string = round(empty_string_count / total_count * 100, 2)

    return {
        'empty_string_count': empty_string_count,
        'taxa_empty_string': taxa_empty_string
    }

def get_flag_analysys(df, target_column_name):
    total_count = df.count()
    flag_count = df.filter(f'{target_column_name} != 1 AND {target_column_name} != 0').count()
    taxa_flag = round(flag_count / total_count * 100, 2)

    return {
        'flag_count': flag_count,
        'taxa_flag': taxa_flag
    }
```

```python
"""
Carrega base de dados
"""

df = spark.read.format('delta').table('workspace.default.churn_modelling')
```

```python
"""
analises de data quality
"""

print( '-' * 10 + 'CONTAGEM DE NULOS' + '-' * 10)
c = 0
for col in df.columns:
    c += 1
    analysis = get_null_analysis(df, col)
    print(f"\033[34m{col}\033[0m: {analysis['null_count']} ({analysis['taxa_null']}%)")
print(f'total {c}\n\n')

tabelas_numericas = ('Tenure', 'CreditScore', 'EstimatedSalary', 'Age', 'NumOfProducts')
print( '-' * 10 + 'CONTAGEM DE NEGATIVOS (CAMPOS NUMÉRICOS APENAS)' + '-' * 10)
c = 0
for col in df.columns:
    if col not in tabelas_numericas:
        continue
    c += 1
    analysis = get_negative_analysis(df, col)
    print(f"\033[34m{col}\033[0m: {analysis['negative_count']} ({analysis['taxa_negative']}%)")
print(f'total {c}\n\n')

tabelas_de_texto = ('Surname', 'Geography', 'Gender')
print( '-' * 10 + 'CONTAGEM DE STRINGS VAZIAS (CAMPOS TEXTUAIS APENAS)' + '-' * 10)
c = 0
for col in df.columns:
    if col not in tabelas_de_texto:
        continue
    c += 1
    analysis = get_empty_string_analysis(df, col)
    print(f"\033[34m{col}\033[0m: {analysis['empty_string_count']} ({analysis['taxa_empty_string']}%)")
print(f'total {c}\n\n')

tabelas_de_flag = ('IsActiveMember', 'Exited')
print( '-' * 10 + 'CONTAGEM DE NÃO FLAG (CAMPOS DE FLAG APENAS)' + '-' * 10)
c = 0
for col in df.columns:
    if col not in tabelas_de_flag:
        continue
    c += 1
    analysis = get_flag_analysys(df, col)
    print(f"\033[34m{col}\033[0m: {analysis['flag_count']} ({analysis['taxa_flag']}%)")
print(f'total {c}')
```
